```python
# mqtts方法

import ssl, time
import paho.mqtt.client as mqtt

c = mqtt.Client(
    client_id="MQTTS-Client",
    callback_api_version=mqtt.CallbackAPIVersion.VERSION2,
)

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
c.tls_set_context(ctx)

c.connect("open.xrak.xyz", 446, 60)
c.loop_start()
i = 0
try:
    while True:
        c.publish("demo/ping", f"hello {i}", qos=0)
        i += 1
        time.sleep(3)
except KeyboardInterrupt:
    pass
finally:
    c.loop_stop()
    c.disconnect()
```

```python
# wss方法
import ssl, time
import paho.mqtt.client as mqtt

c = mqtt.Client(
    client_id="WSS-Client",
    transport="websockets",
    callback_api_version=mqtt.CallbackAPIVersion.VERSION2,
)
c.ws_set_options(path="/mqtt/wss")
c.tls_set(tls_version=ssl.PROTOCOL_TLS_CLIENT)
c.connect("open.xrak.xyz", 443, 60)
c.loop_start()
i = 0
try:
    while True:
        c.publish("demo/ping", f"hello {i}", qos=0)
        i += 1
        time.sleep(3)
except KeyboardInterrupt:
    pass
finally:
    c.loop_stop()
    c.disconnect()
```
